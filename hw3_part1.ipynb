## Question 2.1

```python
import numpy as np
import matplotlib.pyplot as plt

df = np.loadtxt('D2z.txt')
x = y = np.arange(-2, 2.1, 0.1)
xx, yy = np.meshgrid(x, y)
test = np.column_stack((xx.flatten(), yy.flatten()))

def euclidean(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

distances = np.array([[euclidean(j, i) for i in df[:, :2]] for j in test])
nn = np.argmin(distances, axis=1)
predictions = df[nn, 2]
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df[df[:, 2] == 0][:, 0], df[df[:, 2] == 0][:, 1], c='blue', marker='s', s=30, label='Label 0')
plt.scatter(df[df[:, 2] == 1][:, 0], df[df[:, 2] == 1][:, 1], c='red', marker='+', s=30, label='Label 1')
plt.scatter(test[predictions == 0][:, 0], test[predictions == 0][:, 1], c='blue', marker='o', s=10, label='Predicted 0')
plt.scatter(test[predictions == 1][:, 0], test[predictions == 1][:, 1], c='red', marker='o', s=10, label='Predicted 1')
plt.title('1NN')
plt.legend()
plt.show()
```
